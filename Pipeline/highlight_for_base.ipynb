```python
import os
import numpy as np
import pandas as pd

from process_audio import extract_audio, preprocess_audio
from process_video import preprocess_video_every_3_seconds

from tensorflow.keras.models import load_model


import numpy as np
```

```python
video_path = "../pang.mp4"
```

```python
video = preprocess_video_every_3_seconds(video_path, (256, 256), 3)
```

```python
video.shape
```

```python
video_model = load_model("../video_model.h5")
```

```python
res = []

for data in video:
    reshaped_input = np.reshape(data, (-1, 9, 256, 256, 1))
    pred = video_model.predict(reshaped_input)
    res.append(pred)

res[:5]
```

```python
final_predictions = np.argmax(res, axis=1)
final_predictions
```
